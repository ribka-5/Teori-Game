```python
import numpy as np
import pandas as pd

# Baca matriks payoff dari file CSV
def read_payoff_matrix_from_csv(file_path):
    try:
        matrix_data = pd.read_csv(file_path, header=None)
        matrix = matrix_data.to_numpy()
        return matrix
    except Exception as e:
        raise Exception("Gagal membaca file CSV: " + str(e))

def eliminate_rows(matrix_baris, matrix_kolom):
    eliminated_rows = []

    for i in range(matrix_baris.shape[0]):
        keep_row = True

        for j in range(matrix_baris.shape[0]):
            if i != j:
                if all(matrix_baris[i] <= matrix_baris[j]):
                    keep_row = False
                    break

        if not keep_row:
            eliminated_rows.append(matrix_baris[i])

    eliminated_row_indices = []
    for row in eliminated_rows:
        for i, row_kolom in enumerate(matrix_baris):
            if np.array_equal(row, row_kolom):
                eliminated_row_indices.append(i)

    matrix_baris = np.delete(matrix_baris, eliminated_row_indices, axis=0)
    matrix_kolom = np.delete(matrix_kolom, eliminated_row_indices, axis=0)

    return matrix_baris, matrix_kolom

def eliminate_columns(matrix_baris, matrix_kolom):
    eliminated_columns = []

    for j in range(matrix_kolom.shape[1]):
        keep_column = True

        for i in range(matrix_kolom.shape[1]):
            if i != j:  # Jangan bandingkan kolom dengan dirinya sendiri
                if all(matrix_kolom[:, j] <= matrix_kolom[:, i]):
                    keep_column = False
                    break

        if not keep_column:
            eliminated_columns.append(matrix_kolom[:, j])

    eliminated_column_indices = []
    for col in eliminated_columns:
        for i, col_baris in enumerate(matrix_kolom.T):
            if np.array_equal(col, col_baris):
                eliminated_column_indices.append(i)

    matrix_baris = np.delete(matrix_baris, eliminated_column_indices, axis=1)

    matrix_kolom = np.delete(matrix_kolom, eliminated_column_indices, axis=1)

    return matrix_baris, matrix_kolom

def iterarif_NE(matrix_baris, matrix_kolom):
    previous_row_shape = matrix_baris.shape
    previous_column_shape = matrix_kolom.shape

    while True:
        matrix_baris, matrix_kolom = eliminate_rows(matrix_baris, matrix_kolom)
        matrix_baris, matrix_kolom = eliminate_columns(matrix_baris, matrix_kolom)

        if matrix_baris.shape == previous_row_shape and matrix_kolom.shape == previous_column_shape:
            break

        previous_row_shape = matrix_baris.shape
        previous_column_shape = matrix_kolom.shape

    return matrix_baris, matrix_kolom

matrix_baris, matrix_kolom = iterarif_NE(matrix_baris, matrix_kolom)

if __name__ == "__main__":
    try:
        # Ganti 'matriks_payoff_baris.csv' dan 'matriks_payoff_kolom.csv' dengan nama file CSV yang berisi matriks baris dan kolom Anda
        file_path_baris = "teo.csv"
        file_path_kolom = "teo.csv"

        matrix_baris = read_payoff_matrix_from_csv(file_path_baris)
        matrix_kolom = read_payoff_matrix_from_csv(file_path_kolom)

        matrix_baris, matrix_kolom = iterarif_NE(matrix_baris, matrix_kolom)

        print("Matriks Baris setelah eliminasi:")
        print(matrix_baris)
        print("Matriks Kolom setelah eliminasi:")
        print(matrix_kolom)
    except Exception as e:
        print("Terjadi kesalahan:", str(e))
```
